```python
from dataclasses import dataclass

@dataclass
class Current:
    name: str
    freq: float
    mag_dB: float
    phase_deg: float
    dB_3_BW_freqs: tuple[float, float]
```

```python
A = Current(
    name="A",
    freq=6.7500841e6,
    mag_dB=4.32,
    phase_deg=-74.47,
    dB_3_BW_freqs=(6.7207017e6, 6.8182424e6)
)
```
