```python

import cmath
from math import pi


freq = 6.78*1e6
w = 2*pi*freq
Vdc = 350
L = 2.5*1e-6
C = 250*1e-12

Z_res = 1j*w*L + 1/(1j*w*C)
R_load_ohm = 120
Z_res_mag_ohm = abs(Z_res+R_load_ohm) 
Z_res_phase_deg = (180/pi)*cmath.phase(Z_res+R_load_ohm)
```

```python
print(f"Resonant Impedance: {Z_res_mag_ohm:.2f} ohms, Phase: {Z_res_phase_deg:.6f}°")
T_sec = 1/freq
delay_sec = 3.7*1e-9
print(f"Period: {T_sec*1e9:.6} nanoseconds")
print(f"Delay: {delay_sec*1e9:.6} nanoseconds")
print(f"Delay as Degrees: {delay_sec/T_sec*360}°")
```
